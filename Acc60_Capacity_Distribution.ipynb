Estimated Varta60 Capacity Distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.DataFrame({
"Accu60":[21.013,
20.600,
20.134,
21.077,
21.124,
20.254,]
})
```

```python
chart=sns.displot(df, 
                  kind='kde',
                  height=6,
                  aspect=1.5,
                  fill=True,color='green',
                  lw=4,
                  edgecolor='green',
                 )
chart.fig.suptitle('Accu60 Capacity Distribution Plot')
chart.set_xlabels('Capacity (mAh) over 3.0V - 4.2V')
chart.set_ylabels('Probability Density')
chart.legend.remove()
plt.grid()
plt.show()
```

```python
import numpy as np
# Number of bootstrap samples
n_bootstrap_samples = 1000

# Initialize an array to store bootstrap sample means
bootstrap_sample_means = np.zeros(n_bootstrap_samples)

# Perform bootstrapping
for i in range(n_bootstrap_samples):
    # Randomly sample with replacement from the original data
    bootstrap_sample = np.random.choice(df.Accu60, size=len(df), replace=True)
    # Calculate the mean of the bootstrap sample
    bootstrap_sample_means[i] = np.mean(bootstrap_sample)
```

```python
import matplotlib.pyplot as plt

# Plot the probability distribution of the bootstrap sample means
plt.hist(bootstrap_sample_means, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Bootstrap Sample Means Distribution')
plt.show()
```

```python
chart=sns.displot(bootstrap_sample_means, 
                  kind='kde',
                  height=6,
                  aspect=1.5,
                  fill=True,color='green',
                  lw=4,
                  edgecolor='green',
                 )
chart.fig.suptitle('Accu60 Capacity Distribution Plot')
chart.set_xlabels('Capacity (mAh) over 3.0V - 4.2V')
chart.set_ylabels('Probability Density')
#chart.legend.remove()
plt.grid()
plt.show()
```
