```python
import pandas as pd
import numpy as np
df = pd.read_csv("Lake-Height-Data.txt",
                 sep=",",header=0, 
                 dtype={'CHR_ID': object, 
                         'ELEVATION': np.float64, 
                         'DATUM_ADJ':object},
                         usecols=['ELEVATION','READ_DATE'],
                         parse_dates=[1]
                )

df['YEAR'] = df['READ_DATE'].dt.year
df.set_index("READ_DATE",inplace=True)
df.to_csv("temp.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.relplot(x=df.index, y='ELEVATION',data=df)
```

```python
# palette = sns.color_palette("mako_r", 23)
palette = sns.color_palette("husl", 23)
sns.lineplot(x=df.index, y='ELEVATION',data=df,palette=palette,hue="YEAR",linewidth = 3)
```
